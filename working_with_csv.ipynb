opening a csv file from an url

```python
import requests
import pandas as pd
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"   # gitHub link of dataset
headers = {"User-Agent": "Mozilla/5.0","Accept": "text/csv"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)
df = pd.read_csv(data)
print(df.head())

### requests and headers are used when we want to communicate with a web server and control how our request looks to that server...
```

```python

pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level']) # usecols helps in showing those which i have need and input to dataset
```

handling dates

```python
pd.read_csv('ipl_data.csv',parse_dates=['date']).info()
```

converters

```python
# function for converters
def rename(name):
  if name=="Royal Challengers Bangalore":
    return "RCB"
  else:
    return name

pd.read_csv('ipl_data.csv',converters={'team1':rename})
```

na_values parameter

```python
pd.read_csv('aug_train.csv',na_values=['Male']) # used to remove non requuired symbols from any column
```
